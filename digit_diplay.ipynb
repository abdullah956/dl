```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(f"Training data shape (rows, columns): {x_train.shape}")
print(f"Test data shape (rows, columns): {x_test.shape}")
```

```python
plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(1, 10, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(y_train[i])
    plt.axis('off')
plt.show()
```

```python
def display_digit(digit):
    indices = np.where(y_train == digit)[0]
    if len(indices) > 0:
        index = indices[0]
        plt.imshow(x_train[index], cmap='gray')
        plt.title(f"Digit: {digit}")
        plt.show()
    else:
        print("Digit not found in the dataset.")
```

```python
digit = int(input("Enter the digit (0-9) you want to see: "))
```

```python
if 0 <= digit <= 9:
    display_digit(digit)
else:
    print("Invalid input. Please enter a digit between 0 and 9.")
```
